## Define target variable (y) and feature variables (x)

```python
df.head()
```

```python
X = df.drop("Class",axis=1)
```

```python
y = df['Class']
```

```python
X.head()
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import Normalizer
norm = Normalizer()

norm.fit(X)
X = pd.DataFrame(norm.transform(X))
```

```python
cols = [f"V{i}" for i in range(1,101)]
X.columns = cols
```

```python
X.head()
```

```python
plt.plot(X.iloc[1]) # Hill
```

```python
plt.plot(X.iloc[2]) # Hill
```

```python
plt.plot(X.iloc[0]) # Valley
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(X,y,test_size=0.2,random_state=23)
```

```python
train_x.shape, test_x.shape, train_y.shape, test_y.shape
```

## Modeling

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=23)
clf.fit(train_x,train_y)
```

```python
clf.score(test_x,test_y)
```

## Model Evaluation

```python
y_preds = clf.predict(test_x)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(test_y,y_preds))
```

```python
confusion_matrix(test_y,y_preds)
```

## Predictions

```python
rd_sample = train_x.sample(1)
rd_sample
```

```python
rd_sample_pred = clf.predict(rd_sample)
rd_sample_pred
```

Here, it predicts a zero (0), so, let's check the result by plotting the points

```python
plt.plot(rd_sample.T)
```

So, our model was able to correctly predict, that these data points form a valley

## Explanation

The Random Forest Classifier was employed to predict whether a set of data points forms a hill or a valley in the Hill Valley Project. The model demonstrated high performance, as evident from the classification report:

* **Accuracy**: 
The model achieved an impressive accuracy of 99%, indicating its ability to correctly classify the majority of data points.

* **Precision and Recall**: 
Both classes (0 for valley and 1 for hill) exhibit high precision (99% for class 0 and 98% for class 1) and recall (98% for class 0 and 99% for class 1). These metrics highlight the model's capability to accurately identify instances of both hills and valleys while minimizing false positives and false negatives.

* **F1-Score**: 
The F1-score, which balances precision and recall, is exceptionally high for both classes (99% for class 0 and 99% for class 1). This underscores the robustness of the model in providing accurate predictions.

* **Observations**:
No significant challenges or anomalies were observed during the model evaluation. The model performs consistently across both classes, indicating a strong generalization to new data.
